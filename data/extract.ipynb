```python
import pandas as pd
import os
```

```python
raw_path = os.path.abspath(os.path.join(os.getcwd(), "../court-iq/data/raw"))

data = {
    file.replace(".csv",""): pd.read_csv(os.path.join(raw_path, file))
    for file in os.listdir(raw_path) if file.endswith(".csv")
}

print(data.keys())
```

```python
raw_df = pd.concat(data.values(), ignore_index=True)

raw_df
```
